## indexing according to relaese date

```python
a1.set_index("release_date",inplace=True)
a1.index=pd.to_datetime(a1.index)
a1.head(5)
```

## changing time from milliseconds to seconds using anonymous function and then showing the top rows

```python
a1["duration"]=a1["duration_ms"].apply(lambda x:round(x/1000))
a1.drop("duration_ms",inplace=True,axis=1)
a1.duration.head()
```

## making samples of dataframe using sample function

```python
sample=a1.sample(int(0.05*len(a1)))
print(sample)
print(len(sample))
```

## showing relation between loudness and energy

```python
plt.figure(figsize=(5,6))
sns.regplot(data=sample,y="loudness",x="energy",color="c").set(title="loudness vs energy correlation")
```

## this graph shows how energy is associated with loudness,  how much energy  is generated with a specific loudness 

## graph showing relationship between popularity and liveness of songs

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample,y="popularity",x="liveness",color="m").set(title="popularity vs liveness correlation")
```

## this graph shows which songs are popular according to their live telecasting

## showing year of songs according to their release date 

```python
a1['dates']=a1.index.get_level_values('release_date')
a1.dates=pd.to_datetime(a1.dates)
years=a1.dates.dt.year
years
```

## making distplot(density + histogram) on years column

```python
sns.distplot(years).set(title="density of songs per year")
```

## showing duration of songs year-wise

```python
total_dur=a1.duration
fig_dims=(20,7)

fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dur,ax=ax,errwidth=True).set(title="year vs duration")
plt.xticks(rotation=60)
```

## line plot for year and duration of songs

```python
total_dur=a1.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dur,ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```

## this graph  shows the duration of songs according to date with an interval of  20 years

## showing dyration of songs according to genre of songs

## showing heatmap between variables 

```python

corr_df=a1.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(10,6))
heatmap=sns.heatmap(corr_df,annot=True,vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="black")
heatmap.set_title("correaltion heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=67)
```

## this graph shows the correlation between various variables of various columns
